```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt, matplotlib.ticker as ticker
sns.set(style="whitegrid", font_scale=1.25)

def parse_string(string):
    string = string.split(" ")
    hot = float(string[-1])
    canopy = float(string[-2])
    if string[0] in ["Woden", "Canberra", "Molonglo", "Weston", "Paddys"]:
        district = " ".join(string[:2])
        suburb = " ".join(string[2:-2])
    else:
        district = string[0]
        suburb = " ".join(string[1:-2])
    return [suburb, canopy, hot]    

data = pd.read_csv("trees.txt", header=None, names=["string"])
df = pd.DataFrame(columns=["suburb", "canopy", "heat"])

for i in data.index:
    df.loc[i] = parse_string(data.at[i, "string"])

fig, ax = plt.subplots(
    tight_layout=True,
    figsize=(13.5, 24)
)

sns.barplot(x="canopy", y="suburb", ax=ax, data=df, palette="Greens_r")
plt.title("ACT canopy coverage", font={"size":"large", "weight":"bold"})
ax.xaxis.set_major_formatter(ticker.PercentFormatter(100, 0))
plt.xlabel("Area with trees 3 metres or taller (2020 data)")
plt.ylabel(None)
plt.savefig("canopy.jpg")

fig, ax = plt.subplots(
    tight_layout=True,
    figsize=(13.5, 24)
)

sns.barplot(x="heat", y="suburb", ax=ax, data=df.sort_values("heat", ascending=False), palette="Reds_r")
plt.title("ACT heat vulnerability", font={"size":"large", "weight":"bold"})
ax.xaxis.set_major_formatter(ticker.PercentFormatter(100, 0))
plt.xlabel("Area vulnerable to urban heat (2020 data)")
plt.ylabel(None)
plt.savefig("heat.jpg")

df.set_index("suburb", inplace=True)
```
